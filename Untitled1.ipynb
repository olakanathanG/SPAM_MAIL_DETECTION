```python
import pandas as pd
import nltk
```

**import** **dataet**

```python
data=pd.read_csv("/content/spam.csv",encoding="ISO-8859-1")
```

```python
data.head()
```

```python
data=data.drop('Unnamed: 2',axis=1)
```

```python
data.columns=['label','text']
```

**PUNCTUATION**

```python
import string
string.punctuation
```

```python
def romove_punct(text):
  data="".join([char for char in text  if char not in string.punctuation])
  return text
```

```python
data['labels_text']=data['text'].apply(lambda y: romove_punct(y))
data.head()
```

**TOKENIZE**

```python
import re
def tokenizer(text):
  token=re.split('\W+',text)
  return  token
data['token_text']=data['labels_text'].apply(lambda x: tokenizer(x.lower()))
```

```python
data.head()
```

***STOPWORD***

```python
import nltk
nltk.download("corpus")
nltk.download('stopwords')

model=nltk.corpus.stopwords.words('english')
```

```python
def stop_word(token_text):
  stop=[word for word in token_text if word not in model]
  return stop

data['stopword_text']=data['token_text'].apply(lambda x: stop_word(x))
```

```python
data.head()
```

**VECTORIZER**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
model_1=TfidfVectorizer()
model_1
```

```python
mo=model_1.fit_transform(data['stopword_text'].astype(str))
```

```python
x=pd.DataFrame(mo.toarray())
```

```python
x.head()
```

**RANDOM** **FORET** **CLASSIFIER**

```python
from sklearn.ensemble import RandomForestClassifier
model_2=RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,data['label'],test_size=0.2)
```

```python
fit=model_2.fit(x_train,y_train)
```

```python
y_pred=model_2.predict(x_test)
```

```python
from sklearn.metrics import precision_score,recall_score
```

```python
predictionScore=precision_score(y_test,y_pred,pos_label='spam')
recall=recall_score(y_test,y_pred,pos_label='spam')
```
